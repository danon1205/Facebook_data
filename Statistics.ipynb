```python
from pandas import read_csv
import pandas as pd

dataset = read_csv('dataset_Facebook.csv',";")
dataset.dropna(inplace = True)
```

```python
#Вычисление максимальных значений всего массива данных 

print(dataset.max(numeric_only = True))
```

```python
#Вычисление минимальных значений всего массива данных

print(dataset.min(numeric_only = True))
```

```python
#Вычисление средних арифметических всего массива данных 

print(dataset.mean(numeric_only = True))
```

```python
#Вычисление мод всего массива данных(результат в виде таблицы, так как мода может быть не единственной)

print (dataset.mode(numeric_only = True))
```

```python
#Вычисление медиан всего массива данных

print (dataset.median(numeric_only = True))
```
